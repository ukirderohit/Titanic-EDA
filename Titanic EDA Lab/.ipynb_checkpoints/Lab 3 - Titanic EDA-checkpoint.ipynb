Titanic Dataset - train.csv

```python
#Read CSV
df = pd.read_csv("train.csv")
```

```python
df
```

## For Variable - PassengerID

1) To Check is the variable categorical or continuous??

```python
df.PassengerId.value_counts()
```

<b>From this you can see that PassengerId is continuous

2) Are there missing values?

```python
df[df.PassengerId.isnull()]
```

<b> PassengerId has no Missing values

3) Min, Max, Mean, and Standard Deviation of the continuous variables.

```python
df.describe()
```

<b> From Here you can see that for PassengerID <br>
<b> Min :- 1.00 <br>
<b> Max :- 891.00 <br>
<b> Mean :- 446.00 <br>
<b> Standard Deviation :- 257.3538

4) Histograms describing the distribution of the variable.

```python
df.PassengerId.hist()
```
